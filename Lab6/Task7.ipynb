```
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt 
```

```
df_country = pd.read_csv("https://raw.githubusercontent.com/sagihaider/CE888_2021/main/Lab_6/Data/Country-data.csv")
df_dictionary = pd.read_csv("https://raw.githubusercontent.com/sagihaider/CE888_2021/main/Lab_6/Data/data-dictionary.csv")

```

```
X = df_country.iloc[:,1:]
X
```

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_std = sc.fit_transform(X)
```

**Apply PCA with n_components=3 on X_train_std data**

```
pca = PCA(n_components=9)
X_train_pca = pca.fit_transform(X_train_std)
```

```
#X_train_pca
```

K_means clustering to cluster the data

```
from sklearn import cluster, datasets, metrics
km=cluster.KMeans(n_clusters=3)
km.fit(X_train_pca)
label = km.labels_
```
